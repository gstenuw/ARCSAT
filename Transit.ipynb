```python
import batman
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-0.15, 0.15, 1000)

params = batman.TransitParams()
params.t0 = 0                         # time of transit center
params.per = 3.941534                 # orbital period in days
params.rp = 0.132                     # planet radius (Rp/Rs)
params.a = 11.4                       # semi-major axis (a/Rs)
params.inc = 88.8                     # inclination in degrees
params.ecc = 0.0                      # assumed circular orbit
params.w = 90.0                       # argument of periastron
params.u = [0.3, 0.2]                 # limb darkening coefficients
params.limb_dark = "quadratic"       # limb darkening model
#Data from https://exoplanetarchive.ipac.caltech.edu/overview/xo-1b
m = batman.TransitModel(params, t)
flux = m.light_curve(params)

plt.plot(t, flux, label="XO-1 b")
plt.xlabel("Time from mid-transit")
plt.ylabel("flux")
plt.title("XO-1 b Transit Light Curve")
plt.grid(True)
plt.legend()
plt.show()
```
